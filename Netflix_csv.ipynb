and Movies available on Netflix till 2021.

This dataset is collected from Flixable which is a third-party 
Netflix search engine, and available on Kaggle website for free.

##### Q9. For 'House of Cards', what is the Show Id and Who is the Director of this show ?

```python
netflix[netflix.Title =='House of Cards'][['Show_Id','Director']]
#or 
netflix[netflix.Title.isin(['House of Cards'])][['Show_Id','Director']]
# or 
netflix[netflix.Title.str.contains('House of Cards')][['Show_Id','Director']]
```

##### Q10.In which year highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
netflix['Date_N'] = pd.to_datetime(netflix['Release_Date'], format = 'mixed')   # dtype of Release_Date has changed to datetime
netflix['Date_N'].dt.year.value_counts().plot(kind='bar')
plt.show()
```

##### Q11. How many Movies & TV Shows are in the dataset ? Show with Bar Graph.

```python
netflix.Category.value_counts().plot(kind ='bar')
# or
netflix.groupby('Category').Category.count().plot(kind = 'bar')
```

##### Q12. Show all the Movies that were released in year 2010?

```python
# to create new Year column : it will consider only year as value
netflix['Year'] = netflix['Date_N'].dt.year
netflix[(netflix['Category'] == 'Movie')&(netflix['Year']==2010)]
#or
t1 = pd.to_datetime('2010-01-01')
t2 = pd.to_datetime('2011-01-01')
netflix[(netflix['Category'] == 'Movie') & (netflix['Date_N']>= t1) & (netflix['Date_N']< t2)]
```

##### Q13. Show only the Titles of all TV Shows that were released in India only?

```python
netflix.loc[(netflix.Country=='India') & (netflix.Category == 'TV Show'),['Title']]
```

##### Q14. Show Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix ?

```python
netflix[netflix.Director.isin(netflix.Director.value_counts().head(10).index)]
```

##### Q15.  Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom" ?

```python
netflix.loc[(netflix.Category == 'Movie') & (netflix.Type == 'Comedies') | (netflix.Country == 'United Kingdom')]
```

##### Q16. In how many movies/shows, Tom Cruise was cast ?

##### Q18. How many Movies got the 'TV-14' rating, in Canada ?

```python
netflix.loc[(netflix.Rating == 'TV-14') & (netflix.Country == 'Canada') & (netflix.Category == 'Movie')].shape
```

##### Q19. How many TV Show got the 'R' rating, after year 2018 ?

```python
netflix.loc[(netflix.Rating == 'R') & (netflix.Category == 'TV Show') & (netflix.Year > 2018)]
# or
netflix.loc[(netflix.Rating == 'R') & (netflix.Category == 'TV Show') & (netflix.Date_N >= pd.to_datetime('1/1/2019'))]
```

##### Q20. What is the maximum duration of a Movie/Show on Netflix ?

```python
netflix[['Minutes','Unit']] = netflix['Duration'].str.split(' ', expand = True)
netflix.groupby('Unit').get_group('min').Minutes.astype(int).max()
```

##### Q21. Which individual country has the Highest No. of TV Shows ?

```python
netflix.groupby(['Category']).get_group('TV Show').Country.value_counts()
```

##### Q22. How can we sort the dataset by Year ?

```python
netflix.sort_values("Year")
# or
netflix.sort_values('Date_N')
```
