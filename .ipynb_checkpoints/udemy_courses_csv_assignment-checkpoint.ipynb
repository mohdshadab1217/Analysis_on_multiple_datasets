##### 6. Check For Duplicate Data and Drop Them?

```python
udemy[udemy.duplicated()]
udemy.drop_duplicates(inplace = True , ignore_index = True)
udemy
```

##### 7. What are all different subjects for which Udemy is offering courses ?

```python
udemy.subject.unique()
```

##### 8. Find Out Number of Courses Per Subjects?

```python
udemy.subject.value_counts()
```

##### 9. Show all the courses which are Free of Cost?

```python
udemy[udemy.is_paid == False]
```

##### 10. Show all the courses which are Paid?

```python
udemy[udemy.is_paid]
```

##### 11. Display The Count of Paid and Free Courses?

```python
udemy.is_paid.value_counts()
```

##### 12. For Which Levels, Udemy Courses Providing The Courses?

```python
udemy.level.value_counts()
```

##### 13. Which are Top Selling Courses ?

```python
udemy.sort_values('num_subscribers', ascending = False)
```

##### 14. Which are Least Selling Courses ?

```python
udemy.sort_values('num_subscribers')
```

##### 15. Show all courses of Graphic Design where the price is below 100 ?

```python
udemy[(udemy.subject == 'Graphic Design') & (udemy.price <100)]
```

##### 16. Find Most Popular Course Title?

```python
udemy.loc[udemy.num_subscribers.idxmax(),'course_title']
```

##### 17. Display 10 Most Popular Courses As Per Number of Subscribers?

```python
udemy.sort_values('num_subscribers', ascending = False).head(10)
```

##### 18. Find Total Number of Courses Related To Python?

```python
udemy[udemy.course_title.str.contains('Python', case = False)].shape
```

##### 19. Display 10 Most Popular Python Courses As Per Number of Subscribers?

```python
udemy[udemy.course_title.str.contains('Python', case = False)].sort_values('num_subscribers', ascending = False).head(10)
```

##### 20. What are courses that published in year 2015 ?

```python
udemy['published_timestamp'] = pd.to_datetime(udemy.published_timestamp)
udemy[udemy.published_timestamp.dt.year == 2015]
```

##### 21. Display Category-Wise Count of Posted Subjects [Year Wise] ?

```python
udemy.groupby(udemy.published_timestamp.dt.year).subject.value_counts()
```

##### 22. What are the Max. Number of Subscribers for Each Level of courses ?

```python
udemy.groupby('level').num_subscribers.max()
```
