```python
# How to sort records by player in ascending order for the year 2010?
```

```python
# How to find out Top 5 highest paid salary records for the year 2010 (show only 2 columns i.e. player name and salary)
```

```python
df.groupby('Year').get_group(2010).sort_values('Salary', ascending = False)[:2][['Player','Salary']]
# or 
df[df.Year == 2010].sort_values('Salary', ascending = False)[:2][['Player','Salary']]
df.loc[df.Year == 2010,['Player','Salary']].sort_values('Salary', ascending = False)[:2]
```
