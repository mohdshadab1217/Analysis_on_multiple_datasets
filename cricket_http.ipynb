```python
import pandas as pd
import numpy as np
```

```python
cricket_list = pd.read_html('https://www.espncricinfo.com/rankings/content/page/211271.html')
```

```python
cricket = pd.DataFrame()
for i in cricket_list:
    cricket = cricket._append(i, ignore_index =True)
```
